## Basic Libraries II

Similar to JSON, Python includes a Pickle library. However, in contrast to the JSON format, Pickle is a Python-specific serialization format. The Pickle library provides tools to serialize Python objects, which involves transforming them into a stream of bytes. It also allows you to read these byte streams by deserializing them, transforming them back into their original Python objects.

In contrast to the JSON format, the binary format is usually more compact and, therefore, more efficient.

```python
import numpy as np
data = np.random.rand(10)

import pickle

# Serializing (dumping) the object
with open('data.pkl', 'wb') as f:
    pickle.dump(data, f)

# Deserializing (loading) the object
with open('data.pkl', 'rb') as f:
    loaded_data = pickle.load(f)

print(loaded_data)
```

To work with **Parquet** files, you need either the **pyarrow** or **pandas** library. Parquet is a columnar storage format, meaning that each row represents a sample, and each column represents an attribute. This is a powerful format commonly used as a standard in platforms like **Hugging Face**.

```python
import pandas as pd # if it is not working, simply uncomment the following line
# !pip install pandas

# Creating a DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35]
})

# Writing DataFrame to Parquet file with Pandas
df.to_parquet('data.parquet')

# Reading DataFrame from Parquet file with Pandas
df_loaded = pd.read_parquet('data.parquet')

print(df_loaded)
```

## Basic Libraries II

Let's now try to use them to order the annotations by date

```python
import re
import glob
import os
from datetime import datetime

# Regex pattern, r in front of strings tell python to treat them as raw strings
# we do this so slashes don't get interpret as scaping symbol
pattern = r'(\d{8})_(\d{6})_SN(\d+)_QUICKVIEW_VISUAL_([\d_]+)_([A-Za-z0-9\-_.]+)\.txt' 

annotations = glob.glob('session_4/annotations/*.txt')

# let's create a dictionary where per each annotations we gather the datetime object
ann_datetime = []

for annotation in annotations:

    # extract the file name
    filename = os.path.basename(annotation)
    
    # Search and extract values
    match = re.match(pattern, filename)
    if match:
        date, time, _, _, _ = match.groups()

        # Put them together, e.g. "20240101192856"
        datetime_str = date + time 

        # Parse the string into a datetime object
        datetime_obj = datetime.strptime(datetime_str, "%Y%m%d%H%M%S")

        # Output the datetime object
        print(f"Datetime Object: {datetime_obj}")
        
        ann_datetime.append((filename, datetime_obj))
```

### Question 1

1. How many annotations you have per month and year. Which month has more annotation files.

```python
# First create empty lists to fill later 
mon = []
times = []

for ind in range(len(ann_datetime)): # For every index in the datetime annotation 
    if ann_datetime[ind][1]: #if the index excists
        mon.append(ann_datetime[ind][1].month) # Append the month to the mon list


for months in range(1,13): # For range 1 to 12
    times.append(mon.count(months)) #Count the number of times the month appears and appenmd this to the times list

#The times list now contains each months number of occurence in order. 
#Identify the index with the most entries and add one to find the month number
month = times.index(max(times))+1

#Change the month to a datetime type. Since only the month is neccesaey the year and day values can be set to anything
mydate = datetime(year=1,month= month,day=1)

# print the results
print(f'The number of occurances for each month in order:\n{times}')
print(f"\n{mydate.strftime("%B")} has the most annotations with {max(times)} annotations")
```

```python
# First create an empty dictionary and list to fill later
dictionary = {}
anno_list = []

# Iterate through each month
for months in range(1,13):
    # create a datetime object with the month 
    dating = datetime(1,months,1) 

    # Convert the month to a string 
    month_string = dating.strftime("%B") 

    for ind in range(len(ann_datetime)): #Iterate though the annotations

        #Check If the annotation month matches the month
        if ann_datetime[ind][1].month==months: 

            # Append to the list of annotation
            anno_list.append(ann_datetime[ind][0]) 

    # Add a key of the month to the dictionary with the values of the list of annotation
    dictionary[month_string] = anno_list 

    # Empty the list top be used again for the next month
    anno_list=[] 

#print the dictionary
print(dictionary)

# Print for the month of February as an example. Make sure the date lists the year and then the month as 02
dictionary['February']
```
